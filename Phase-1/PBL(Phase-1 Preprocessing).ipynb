```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df1 = pd.read_csv("PBL-dataset1.csv")
df1
```

```python
df2 = pd.read_csv("PBL-dataset2.csv")
df2
```

```python
df = pd.concat([df1, df2], axis=0)
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
print(df.isnull().sum())
```

```python
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df
```

```python
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())
df['Rating'] = df['Rating'].round(1)
df
```

```python
df['Verified_Purchase'].fillna('Unknown', inplace=True)
df
```

```python
df['Would_Recommend'].fillna('Unknown', inplace=True)
df
```

```python
df['Review_Date'] = pd.to_datetime(df['Review_Date'], format='mixed', dayfirst=True)
df
```

```python
df['Location'] = df['Location'].str.title()
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df = df[df['Would_Recommend'] != 'Unknown']

df['Would_Recommend'] = df['Would_Recommend'].map({
    'No': 0,
    'Maybe': 1,
    'Yes': 2
})
df
```

```python
df['Review_Text']
```

```python
import re

def clean_text(text):
    text = re.sub(r'<.*?>', '', text)   # remove HTML tags
    text = re.sub(r'[^a-zA-Z]', ' ', text)  # remove punctuation & numbers
    text = text.lower()
    return text

df['Review_Text'] = df['Review_Text'].apply(clean_text)
df
```

```python
df.to_csv("PBL_final.csv", index=False)
```
